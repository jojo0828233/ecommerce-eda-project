## Introduction
The purpose of this analysis is to evaluate how different age groups contribute to overall revenue.
By segmenting customers into defined age brackets (18–25, 26–35, 36–50, 50+), we can identify spending patterns and determine which demographic generates the highest sales.
This helps us understand buyer behavior across age ranges and provides valuable insights for targeted marketing strategies and product positioning.


## Methodology  
To analyze revenue performance across different age groups, the following steps were carried out:  

- **Data Cleaning**  
  - Verified the *Age* column for missing values, outliers, or incorrect entries.  
  - Standardized age values to ensure consistency.  

- **Age Binning**  
  - Created categorical bins for age ranges:  
    - 18–25  
    - 26–35  
    - 36–50  
    - 50+  

- **Grouping & Aggregation**  
  - Grouped the dataset by age bins.  
  - Calculated total sales/revenue for each age group.  

- **Visualization**  
  - Plotted a bar chart to compare revenue contributions across age brackets.  
  - Applied clear labels, colors, and axis titles for readability.  

- **Insights Extraction**  
  - Interpreted the chart to identify which age group generated the highest revenue.  
  - Highlighted spending patterns and possible behavioral trends among different demographics.  

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../E-commerce_dataset.csv')
df.head()
```

```python
df['age'] = pd.to_numeric(df['age'], errors='coerce') 
df = df.dropna(subset=['age'])
```

```python
bins = [17, 25, 35, 50, 120] 
labels = ['18-25', '26-35', '36-50', '50+']

df['AgeGroup'] = pd.cut(df['age'], bins=bins, labels=labels)
```

```python
age_group_sales = df.groupby('AgeGroup')['sales'].sum().reset_index()
print(age_group_sales)
```

```python
plt.figure(figsize=(8,6))
plt.bar(age_group_sales['AgeGroup'], age_group_sales['sales'], color=['#4C72B0','#55A868','#C44E52','#8172B3'])
plt.title('Revenue by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Total Revenue')
plt.show()
```

## Conclusion

The analysis reveals a clear upward trend in revenue contribution across age brackets, with older groups spending more. The **50+ age group** generated the highest total revenue, followed by **36–50**, **26–35**, and **18–25**. This pattern suggests that purchasing power increases with age, likely due to greater financial stability and disposable income among older consumers. These insights can inform targeted marketing strategies, product positioning, and user experience design tailored to high-value age segments.
