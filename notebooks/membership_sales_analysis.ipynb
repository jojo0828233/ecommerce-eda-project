# introduction
The objective of this analysis is to evaluate the relationship between customer membership status (Normal vs. Premium) and financial performance. By isolating the sales data attached to different membership tiers, we aim to understand which customer segment drives the most revenue and determine the value proposition of the Premium membership program.

# Methodology
1. **Data Loading**: The dataset was imported using Pandas.

2. **Filtering**: The data was filtered to exclude rows where membership was labeled "Not Logged In" to ensure the comparison focuses strictly on established membership tiers.

3. **Grouping**: The data was grouped by the membership column.

4. **Aggregation**: We calculated the Total Sales (sum) to understand total volume. Note: The code below also adds a calculation for Average Sales (mean) to evaluate transaction value, which was a requested task not fully calculated in the original cells.

5. **Visualization**: A bar chart was generated to visually compare the revenue contribution of each group.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("..\E-commerce_dataset.csv")
df.head()
```

# Total sales



```python
total_sales = df.groupby("membership")["sales"].sum()
pd.options.display.float_format = '{:,.2f}'.format
df = df[df["membership"] != "Not Logged In"]
total_sales
```

```python
total_sales.plot(kind="bar")
plt.title("Total Sales by Membership")
plt.xlabel("Membership")
plt.ylabel("Total Sales")
plt.ticklabel_format(style='plain', axis='y')
plt.show()
```

# Average sales

```python
avg_sales = df.groupby("membership")["sales"].mean()
pd.options.display.float_format = '{:,.2f}'.format
df = df[df["membership"] != "Not Logged In"]
avg_sales
```

```python
avg_sales.plot(kind="bar")
plt.title("Average Sales by Membership")
plt.xlabel("Membership")
plt.ylabel("Average Sales")
plt.ticklabel_format(style='plain', axis='y')
plt.show()
```

# Conclusion
The analysis clearly indicates that Premium membership is a critical driver of revenue for this e-commerce platform. While Normal members contribute a healthy $22.7M, the Premium tier generates the bulk of the sales at $48.4M
