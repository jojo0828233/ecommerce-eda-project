## Introduction

Identifying language usage patterns helps reveal which audience segments are most active, how diverse the user population is, and whether the platform effectively supports multilingual users. The purpose of this task is to visualize and compare the frequency of each language used by customers, determine the most commonly used languages, and understand the overall linguistic distribution within the dataset.





## Methodology
The dataset was loaded using Pythonâ€™s `pandas` library.  
A value count operation (`value_counts()`) was performed on the **language** column to determine how many times each language appears in the dataset.

To clearly highlight the most dominant languages, a horizontal bar graph was generated using `matplotlib.pyplot`.  
The top two most frequently used languages were visually emphasized using an orange color, while all remaining languages were displayed in blue. This color distinction helps quickly identify the leading languages at a glance.

The steps followed were:
1. Importing required libraries (`pandas` and `matplotlib`).
2. Reading the dataset from the specified CSV file.
3. Counting the occurrences of each language.
4. Sorting and preparing the data for visualization.
5. Plotting a horizontal bar chart with a custom color scheme to highlight the top two languages.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import squarify
```

```python
df = pd.read_csv("C:\\Users\\SHIVAPRASAD\\Desktop\\Project\\ecommerce-eda-project\\E-commerce_dataset.csv")
df.head()
```

```python

print(df["language"].unique())
```

```python
counts = df["language"].value_counts()


colors = ['orange' if i < 2 else 'blue' for i in range(len(counts))]

plt.figure(figsize=(8,5))
plt.barh(counts.index, counts.values, color=colors)
plt.xlabel("Count")
plt.title("Most Used Languages")
plt.show()
```
