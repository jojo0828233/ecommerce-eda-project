```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import squarify

```

```python
df = pd.read_csv("C:\\Users\\SHIVAPRASAD\\Desktop\\Project\\ecommerce-eda-project\\E-commerce_dataset.csv")
df.head()
```

```python

print(df["pay_method"].unique())
```

```python

sales_sum = df.groupby("pay_method")["sales"].sum().reset_index()

# Optional: create custom colors
colors = ["orange" if i < 2 else "skyblue" for i in range(len(sales_sum))]

# Bar plot
sns.barplot(data=sales_sum, x="pay_method", y="sales", palette=colors)

plt.xlabel("Payment Method")
plt.ylabel("Total Sales")
plt.title("Total Sales by Payment Method")
plt.show()

```
